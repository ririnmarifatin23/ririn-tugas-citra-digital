```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

# Folder tujuan untuk menyimpan hasil
output_folder = r"C:\Users\user\tugas"
os.makedirs(output_folder, exist_ok=True)

# Path input gambar (pastikan file ada di folder kerja atau berikan path lengkap)
input_image1 = r"C:\Users\user\tugas\1.png"
input_image2 = r"C:\Users\user\tugas\2.png"

# Baca gambar dalam format grayscale
img1 = cv2.imread(input_image1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(input_image2, cv2.IMREAD_GRAYSCALE)

# Cek apakah gambar berhasil dibaca
if img1 is None or img2 is None:
    print("Gagal membaca salah satu gambar. Cek path-nya.")
else:
    # Threshold untuk ubah jadi biner
    _, binary1 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
    _, binary2 = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)

    # Buat kernel
    kernel = np.ones((5, 5), np.uint8)

    # Proses dilasi
    dilated1 = cv2.dilate(binary1, kernel, iterations=1)
    dilated2 = cv2.dilate(binary2, kernel, iterations=1)

    # Simpan hasilnya
    cv2.imwrite(os.path.join(output_folder, "dilated_1.png"), dilated1)
    cv2.imwrite(os.path.join(output_folder, "dilated_2.png"), dilated2)

    # Tampilkan hasil di notebook
    plt.figure(figsize=(10, 4))

    plt.subplot(2, 2, 1)
    plt.title("Original 1")
    plt.imshow(binary1, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.title("Dilated 1")
    plt.imshow(dilated1, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.title("Original 2")
    plt.imshow(binary2, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.title("Dilated 2")
    plt.imshow(dilated2, cmap='gray')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

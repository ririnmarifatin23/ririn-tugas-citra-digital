```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

# Folder tujuan untuk menyimpan hasil
output_folder = r"C:\Users\user\tugas"
os.makedirs(output_folder, exist_ok=True)

# Path input gambar
input_image1 = os.path.join(output_folder, "1.png")
input_image2 = os.path.join(output_folder, "2.png")

# Baca gambar dalam format grayscale
img1 = cv2.imread(input_image1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(input_image2, cv2.IMREAD_GRAYSCALE)

# Cek apakah gambar berhasil dibaca
if img1 is None or img2 is None:
    print("Gagal membaca salah satu gambar. Periksa path gambar.")
else:
    # Threshold untuk ubah ke gambar biner
    _, binary1 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
    _, binary2 = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)

    # Structuring element (kernel)
    kernel = np.ones((5, 5), np.uint8)

    # Operasi erosi
    eroded1 = cv2.erode(binary1, kernel, iterations=1)
    eroded2 = cv2.erode(binary2, kernel, iterations=1)

    # Simpan hasil ke folder
    cv2.imwrite(os.path.join(output_folder, "eroded_1.png"), eroded1)
    cv2.imwrite(os.path.join(output_folder, "eroded_2.png"), eroded2)

    # Tampilkan hasil di Jupyter Notebook
    plt.figure(figsize=(10, 4))

    plt.subplot(2, 2, 1)
    plt.title("Original 1")
    plt.imshow(binary1, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.title("Eroded 1")
    plt.imshow(eroded1, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.title("Original 2")
    plt.imshow(binary2, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.title("Eroded 2")
    plt.imshow(eroded2, cmap='gray')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
